# Pandas
Pandas can be used to load data from csv files stored in local drive or from url.


The following code assumes the 'xyz.csv' file in the same directory as the python script.

This dataframe does not have any names for the columns. Often the csv files have an extra header row where all the column names are given. We can also add names to the columns while reading the csv file.

````
from pandas import read_csv
url = 'https://raw.githubusercontent.com/swakkhar/MachineLearning/master/pima-indians-diabetes.data.csv'
names = ['num_preg','p_glucose','bp','triceps','2hr_insulin','bmi','dpf','age','class']
data = read_csv(url,names=names)
````

Now lets peek the data from the top 7 rows.

````
data.head()
````

You can notice the column names here. Now its possible to access the columns using the names.

````
data['age']
````

Also a level deeper...

````
data['age'][3]
````

The function shape let us to examine the size of the dataframe or other data types from numpy as well.

````
data.shape
````

The basic data visualization is to plot any column or the dataframe.

````
data['age'].plot()
````

If we want to change the plot size.

````
data['age'].plot(figsize=(15,5))
````

loc is a useful operation on dataframes to select particular row and columns. Here columns are accessible by names/indices

````
part=data.loc[2:7, ['age','bmi']]
part
````

````
data.loc[40:49,'age'].plot(figsize=(15,5))
````

````
data.loc[0:5,'age'].plot.bar()
````

````
data['age'].plot(kind='density')
````

Scatter plot is another way to visualize. It can show correlation between two sets / columns.

````
data.plot.scatter('age','bmi')
````

There is a function info() to get a summary of the types of the data frame and its columns

````
data.info()
````

````
data.tail()
````

````
import matplotlib.pyplot as plt
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True) #autopct is show the % on plot
````

# Data Preprocessing

There are several tasks in data preprocessing
* **Data Cleaning** It involved checking the following:
  * Validity
  * Accuracy
  * Completeness
  * Uniformity
* **Feature Selection** There are several benefits:
  * Reduces Overfitting
  * Improves Performance
  * Reduces training time and mmemory footprint
* **Data Transformation**
  * Rescaling
  * Standardization
  * Normalization

# Encoding
There are several encoding methods.
* Label Encoder
* One-Hot Encoder
* Binary Encoder

````
german.head()
````

````
german['Sex'] = german['Sex'].astype('category')
german['Sex'] = german['Sex'].cat.codes
````

````
german.head()
````

````
german['Saving accounts'].value_counts()
````

````
german['Saving accounts'].value_counts()
````

````
german['Saving accounts'].fillna(method='bfill')
````

````
german.head()
````

````
german['Saving accounts'] = german['Saving accounts'].fillna(method='bfill')
german['Saving accounts'] = german['Saving accounts'].astype('category')
german['Saving accounts'] = german['Saving accounts'].cat.codes

#german['Checking account'].fillna(method='bfill')
#german['Checking account'] = german['Checking account'].astype('category')
#german['Checking account'] = german['Checking account'].cat.codes

#german['Purpose'] = german['Purpose'].astype('category')
#german['Purpose'] = german['Purpose'].cat.codes

#german['Housing'] = german['Housing'].astype('category')
#german['Housing'] = german['Housing'].cat.codes
````
